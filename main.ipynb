```python
def drop_2012(df):
  mask = ['2012' in c for c in df.columns]
  df = df.drop(columns=df.columns[mask])
  return df
```

```python
def winsorizacao_dados(df, string):
  for c in df.columns:
    if string in c :
      df[c] = df[c].clip(lower=df[c].quantile(0.01), upper=df[c].quantile(0.99))
  return df
```

```python
def drop_country_if_less_than_20_companies(df):
  for country in df['Country of Headquarters'].unique():
    mask = (df['Country of Headquarters'] == country)
    if df.loc[mask].shape[0] < 20:
      df = df.drop(index=df.loc[mask].index)
  return df
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score

def regressao_linear(df, coluna_X, coluna_y):
    """
    Realiza a regressão linear entre duas colunas de um dataframe, padroniza as variáveis
    e calcula o R² da regressão.

    Parâmetros:
    df (pd.DataFrame): DataFrame contendo as colunas para regressão.
    coluna_X (str): Nome da coluna que será usada como variável independente.
    coluna_y (str): Nome da coluna que será usada como variável dependente.

    Retorna:
    dict: Coeficiente, intercepto e R² da regressão.
    """

    # Remover observações com NaN nas colunas especificadas
    df_clean = df[[coluna_X, coluna_y]].dropna()

    # Separar variáveis independentes e dependentes
    X = df_clean[[coluna_X]]
    y = df_clean[[coluna_y]]

    # Padronizar as variáveis
    scaler_X = StandardScaler()
    scaler_y = StandardScaler()

    X_scaled = scaler_X.fit_transform(X)
    y_scaled = scaler_y.fit_transform(y)

    # Realizar a regressão linear
    model = LinearRegression()
    model.fit(X_scaled, y_scaled)

    # Obter os coeficientes e intercepto
    coef = model.coef_[0][0]
    intercept = model.intercept_[0]

    # Prever valores e calcular R²
    y_pred = model.predict(X_scaled)
    r_squared = r2_score(y_scaled, y_pred)

    return r_squared#{
        #'coeficiente': coef,
        #'intercepto': intercept,
        #'R2': r_squared
    #}

# Exemplo de uso:
# resultado = regressao_linear(df_usa_2014, 'ESG_2014', 'Qtobin_2014')
# print(resultado)

def regressao_linear2(df,coluna_X1, coluna_X2, coluna_y):
    """
    Realiza a regressão linear entre tres colunas de um dataframe, padroniza as variáveis
    e calcula o R² da regressão.

    Parâmetros:
    df (pd.DataFrame): DataFrame contendo as colunas para regressão.
    coluna_X1 (str): Nome da coluna que será usada como variável independente.
    coluna_X2 (str): Nome da coluna que será usada como variável independente.
    coluna_y (str): Nome da coluna que será usada como variável dependente.

    Retorna:
    dict: Coeficiente, intercepto e R² da regressão.
    """

    # Remover observações com NaN nas colunas especificadas
    df_clean = df[[coluna_X1, coluna_X2,coluna_y]].dropna()

    # Separar variáveis independentes e dependentes
    X = df_clean[[coluna_X1, coluna_X2]]
    y = df_clean[[coluna_y]]

    # Padronizar as variáveis
    scaler_X = StandardScaler()
    scaler_y = StandardScaler()

    X_scaled = scaler_X.fit_transform(X)
    y_scaled = scaler_y.fit_transform(y)

    # Realizar a regressão linear
    model = LinearRegression()
    model.fit(X_scaled, y_scaled)

    # Obter os coeficientes e intercepto
    coef = model.coef_[0]
    intercept = model.intercept_[0]

    # Prever valores e calcular R²
    y_pred = model.predict(X_scaled)
    r_squared = r2_score(y_scaled, y_pred)

    return r_squared#{
        #'coeficiente': coef,
        #'intercepto': intercept,
        #'R2': r_squared
    #}

import statsmodels.api as sm

def regressao_linear_com_pvalue(df, coluna_X, coluna_y):
    """
    Realiza a regressão linear entre duas colunas de um dataframe, padroniza as variáveis
    e calcula o R² da regressão, incluindo o p-value dos coeficientes.

    Parâmetros:
    df (pd.DataFrame): DataFrame contendo as colunas para regressão.
    coluna_X (str): Nome da coluna que será usada como variável independente.
    coluna_y (str): Nome da coluna que será usada como variável dependente.

    Retorna:
    dict: Coeficiente, p-value, intercepto e R² da regressão.
    """

    # Remover observações com NaN nas colunas especificadas
    df_clean = df[[coluna_X, coluna_y]].dropna()

    # Separar variáveis independentes e dependentes
    X = df_clean[[coluna_X]]
    y = df_clean[coluna_y]

    # Adicionar constante para o termo de intercepto
    X = sm.add_constant(X)

    # Realizar a regressão linear
    model = sm.OLS(y, X).fit()

    # Obter os coeficientes, p-value e R²
    coef = model.params[coluna_X]
    intercept = model.params['const']
    p_value = model.pvalues[coluna_X]
    r_squared = model.rsquared

    return {
        'coeficiente': coef,
        'p_value': p_value,
        'intercepto': intercept,
        'R2': r_squared
    }



# Exemplo de uso:
# tabela_estatisticas = gerar_tabela_estatisticas(df_usa_2014, 'ESG_2014', 'Qtobin_2014')
# print(tabela_estatisticas)

```

```python
def regressao_qtobin(df=None):
  for country in df['Country of Headquarters'].unique():
    mask = (df['Country of Headquarters'] == country)
    df_temp = df.loc[mask]
    country_rsquare_ESG_dict[country] = [ regressao_linear(df_temp, x, y) for x, y in list(zip(sorted(ESG_columns), sorted(Q_tobin_columns))) ]
    country_rsquare_E_dict[country] = [ regressao_linear(df_temp, x, y) for x, y in list(zip(sorted(E_columns), sorted(Q_tobin_columns))) ]
    country_rsquare_S_dict[country] = [ regressao_linear(df_temp, x, y) for x, y in list(zip(sorted(S_columns), sorted(Q_tobin_columns))) ]
    country_rsquare_G_dict[country] = [ regressao_linear(df_temp, x, y) for x, y in list(zip(sorted(G_columns), sorted(Q_tobin_columns))) ]

def regressao_roa(df=None):
  for country in df['Country of Headquarters'].unique():
    mask = (df['Country of Headquarters'] == country)
    df_temp = df.loc[mask]
    country_rsquare_ESG_dict[country] = [ regressao_linear(df_temp, x, y) for x, y in list(zip(sorted(ESG_columns), sorted(ROA_columns[0:len(ROA_columns)]))) ]
    country_rsquare_E_dict[country] = [ regressao_linear(df_temp, x, y) for x, y in list(zip(sorted(E_columns), sorted(ROA_columns[0:len(ROA_columns)]))) ]
    country_rsquare_S_dict[country] = [ regressao_linear(df_temp, x, y) for x, y in list(zip(sorted(S_columns), sorted(ROA_columns[0:len(ROA_columns)]))) ]
    country_rsquare_G_dict[country] = [ regressao_linear(df_temp, x, y) for x, y in list(zip(sorted(G_columns), sorted(ROA_columns[0:len(ROA_columns)]))) ]
```

```python
import statsmodels.api as sm
import pandas as pd

def gerar_tabela_estatisticas(X, y):
    """
    Realiza a regressão linear entre arrays e gera uma tabela com as estatísticas do modelo.

    Parâmetros:
    X (array-like): Variável(is) independente(s).
    y (array-like): Variável dependente.

    Retorna:
    pd.DataFrame: Tabela contendo coeficiente(s), desvio padrão, razão t, p-valor, estatística F, R² e R² ajustado.
    """

    # Adicionar constante para o termo de intercepto
    X = sm.add_constant(X)

    # Realizar a regressão linear
    model = sm.OLS(y, X).fit()

    # Inicializar a lista para armazenar as estatísticas
    estatisticas = []

    # Verificar se X tem mais de uma variável independente
    if X.shape[1] > 1:
        # Iterar sobre cada variável independente
        for i in range(1, X.shape[1]):
            estatisticas.append({
                'Estatística': f'Coeficiente X{i}',
                'Valor': model.params[i]
            })
            estatisticas.append({
                'Estatística': f'Desvio padrão X{i}',
                'Valor': model.bse[i]
            })
            estatisticas.append({
                'Estatística': f'Razão t X{i}',
                'Valor': model.tvalues[i]
            })
            estatisticas.append({
                'Estatística': f'P-valor X{i}',
                'Valor': model.pvalues[i]
            })
    else:
        estatisticas.append({
            'Estatística': 'Coeficiente',
            'Valor': model.params[1]
        })
        estatisticas.append({
            'Estatística': 'Desvio padrão',
            'Valor': model.bse[1]
        })
        estatisticas.append({
            'Estatística': 'Razão t',
            'Valor': model.tvalues[1]
        })
        estatisticas.append({
            'Estatística': 'P-valor',
            'Valor': model.pvalues[1]
        })

    # Adicionar estatísticas gerais
    estatisticas.append({
        'Estatística': 'Intercepto',
        'Valor': model.params[0]
    })
    estatisticas.append({
        'Estatística': 'Estatística F (Prob > F)',
        'Valor': model.fvalue
    })
    estatisticas.append({
        'Estatística': 'P-valor Estatística F',
        'Valor': model.f_pvalue
    })
    estatisticas.append({
        'Estatística': 'R squared',
        'Valor': model.rsquared
    })
    estatisticas.append({
        'Estatística': 'R squared adjusted',
        'Valor': model.rsquared_adj
    })

    # Converter a lista de dicionários em um DataFrame
    tabela = pd.DataFrame(estatisticas)

    return tabela

# Exemplo de uso:
# X = np.array([[...], [...], ...])  # Variáveis independentes
# y = np.array([...])  # Variável dependente
# tabela_estatisticas = gerar_tabela_estatisticas(X, y)
# print(tabela_estatisticas)
```

```python
def plot_r2(df=None):
  df_long = df.reset_index().melt(id_vars='index', var_name='Year', value_name='R_squared')

  # Renomeando a coluna 'index' para 'Country'
  df_long.rename(columns={'index': 'Country'}, inplace=True)

  # Plotar o gráfico usando seaborn e matplotlib
  plt.figure(figsize=(14, 8))
  sns.lineplot(data=df_long, x='Year', y='R_squared', hue='Country', palette='tab20', marker='o')

  # Configurações do gráfico
  plt.title('R² por País ao Longo dos Anos')
  plt.xlabel('Ano')
  plt.ylabel('R²')
  plt.grid(True)
  plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
  plt.tight_layout()

  # Mostrar o gráfico
  plt.show()
```
